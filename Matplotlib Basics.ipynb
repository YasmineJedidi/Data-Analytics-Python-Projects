## Plot using a for Loop

```python
plt.title("Gas Prices over Time in $",fontdict={'fontweight':'bold','fontsize':15})

for country in gas.columns:
    if country!="Year":
        plt.plot(gas['Year'], gas[country], marker=".", label=country)
        
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('Year')
plt.ylabel('Gas Price')
plt.legend()

plt.savefig('Gas_price_line_graph',dpi=500)
plt.show()
```

# 3. Pie Chart

```python
fifa.columns
```

```python
plt.title('Preferred Foot for FIFA Players')

left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

lbl=["Left foot","Right foot"]
plt.pie([left,right],labels=lbl,autopct='%.2f%%')
plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
#covert weights to Kg
import re

def convert_weight(weight):
    pattern = re.compile(r'(\d+)')
    if isinstance(weight, str):
        pounds = int(pattern.search(weight).group())
        kg = round(pounds * 0.45359237, 2)
        return kg
    else:
        return None

fifa['Weight_Kg'] = fifa['Weight'].apply(convert_weight)
fifa['Weight_Kg']
```

```python
plt.title('Weight of Professional Soccer Players (Kg)')

light = fifa.loc[fifa.Weight_Kg < 56].count()[0]
light_medium = fifa[(fifa.Weight_Kg >= 56) & (fifa.Weight_Kg < 68)].count()[0]
medium = fifa[(fifa.Weight_Kg >= 68) & (fifa.Weight_Kg < 79)].count()[0]
medium_heavy = fifa[(fifa.Weight_Kg >= 79) & (fifa.Weight_Kg < 91)].count()[0]
heavy = fifa[fifa.Weight_Kg >= 91].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
lbls = ['light','light medium', 'medium', 'medium heavy', 'heavy']

explode = (.4,.2,0,0,.4)
plt.pie(weights, labels = lbls,explode=explode)

plt.show()
```

# 4. Box and Whisker Plot

```python
plt.figure(figsize=(5,8), dpi=100)

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

labels=['Barcelona','Madrid','New England Revolution']
bp = plt.boxplot([barcelona, madrid, revs], labels=labels,patch_artist=True,medianprops={'linewidth': 2})

plt.title('Professional Football Team Comparison')
plt.ylabel('Ovreall Score')
plt.xlabel('Team')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' ) #patch_artist

plt.show()
```
